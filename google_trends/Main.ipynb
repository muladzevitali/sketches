```python
import pandas

from pytrends.request import TrendReq
```

```python
pytrend = TrendReq()
```

```python
kwargs = ["NBA", "Football", "soccer"]
pytrend.build_payload(kw_list=kwargs)
data = pytrend.interest_by_region()
data.sort_values(kwargs, ascending=False)
```
